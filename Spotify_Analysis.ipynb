```python
df_tracks["Duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis=1)
```

```python
df = df_tracks.copy()
```

```python
corr_df = df.drop(["key","mode","explicit","id","name","artists","id_artists","release_date"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot = True, fmt=".1g", vmin=-1,vmax=1,center=0,cmap="inferno",linewidth=1, linecolor="Black")
heatmap.set_title("Correlation Heatmap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
df_tracks = df_tracks.set_index('release_date')
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates, format='mixed')
years=df_tracks.dates.dt.year
```

```python
sns.set_style("white")
plt.figure(figsize = (8, 6))
plt.hist(years,
         bins=87,
         color = "darkseagreen")
plt.title("Number of songs per year", fontsize = 14, weight = "bold")
plt.xlabel("count")
plt.ylabel("years")
sns.despine()
plt.show()
```

```python
total_dr = df_tracks.Duration
fig_dims=(18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years, y=total_dr, ax = ax, errwidth=False).set(title="Year vs. Duraiton")
plt.xticks(rotation=90)
```

```python
total_dr=df_tracks.Duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax = plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title="Year vs. Duration")
plt.xticks(rotation=60)
```
